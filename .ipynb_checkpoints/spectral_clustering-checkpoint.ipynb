```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import scipy.sparse as ss
import scipy.sparse.linalg as ssla
import scipy.linalg as sla
import scipy.stats as stats
from scipy.sparse import csc_matrix
from scipy.sparse import dok_matrix
from scipy.sparse import coo_matrix
from numpy import linalg as LA
from numba import jit, cuda 
```

```python
# image loading and pre-processing into gray scale image and matrix

img = mpimg.imread('3.jpg')
rgb_weights = [0.2989, 0.5870, 0.1140]
newimg = np.dot(img[...,:3], rgb_weights)
plt.imshow(img, cmap=plt.get_cmap("gray"))

#print(newimg)
#rows = len(newimg)
#columns = len(newimg[0])
matimg = np.matrix(newimg)

#Apply SVD to efficiently decrease the size of the image
u,s,v = LA.svd(matimg)
#newimg2 = np.dot(u[:,:160],np.dot(np.diag(s)[:160,:160],v[:160,:]))
newimg2 = newimg
plt.imshow(newimg2)
rows = len(newimg2[:,0])
columns = len(newimg2[0,:])
print(newimg2)
print(rows)
print(columns)
np.shape(newimg2)
```

```python
# Finding the similiarity matrix using 
# k-nearest-neighborhood method.
# reshape check.
shape_org = np.shape(newimg2) # origianl pixel shape
size = np.size(newimg2) # total number of pixels.
newimg2 = np.array(newimg2)
pixarray = np.reshape(newimg2,size) # resahped array 
print(np.size(pixarray))
print(pixarray[0])
newimgbackup = np.reshape(pixarray,shape_org)
plt.imshow(newimgbackup)
```

```python
# pixel position vector should be assigned for each pixel 
x = np.zeros([np.size(pixarray),2]);

for i in range(0,shape_org[0]):
    for j in range(0,shape_org[1]):
        x[shape_org[1]*(i)+j] = [i, j];
```

```python
# Build Weight matrix
# We need to define , sig_f, sig_x, r
sig_f = 0.5
sig_x = 10
r = 5
S = np.size(pixarray)
S = np.size(pixarray)
W = dok_matrix((S,S))
# Choosing a sparse matrix and its type is important in manipulating the matrix.
arraysize=np.size(pixarray)
```

```python
jit(target ="cuda") 
for i in range (0,17160):
    for j in range (i,17160):
        normf = LA.norm(pixarray[i]-pixarray[j])
        normx = (LA.norm(x[i]-x[j]))
        if normx < r:
            W[i,j] = np.exp(-(normf**2/(sig_f**2)))*np.exp(-(normx**2/(sig_x**2)))
            W[j,i] = W[i,j]
```

```python
format(W[1,2], '.5f')
# Improvement required for this algorithm when dealing with huge matrices.
```

```python
# change it into standard eigenvalue problem
# using normalizd Laplacian (Lsym)
D = dok_matrix((S,S))
for k in range (0,17160):
    D[k,k] = np.sum(W[k,:])
```

```python
[w,v] = ssla.eigs(D-W,3,which='SR',sigma = 0.000001, M=D) # Important to choose sigma to be efficiently small. For better result in eigenvecttor.
```
